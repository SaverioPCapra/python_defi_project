```python
import matplotlib.pyplot as plt
```

```python
model = MarketSimulator2D_CLA(50, True)

fig, axs = plt.subplots(2,2,figsize = (10,15))

model.grid = model.create_random_grid()
grid = model.grid.copy()

model.plot_grid(ax = axs[1,0])

model.run_simulation(num_steps=50,
                    P =1,
                    alpha= 0.5,
                    theta= 0.02,
                    initial_mf= 0,
                    mf_volatility= 0,
                    start_new_grid = False)

model.plot_hist(holders_color= "blue", only_holders= False, ax = axs[0,0], title = "Evolution, Fig. 17 (b - 0.5 alpha)")

model.grid = grid
model.plot_grid(ax = axs[1,1])
model.run_simulation(num_steps=50,
                    P =1,
                    alpha= 0.2,
                    theta= 0.02,
                    initial_mf= 0,
                    mf_volatility= 0,
                    start_new_grid = False)

model.plot_hist(holders_color= "black", only_holders= False, ax = axs[0,1], title = "Evolution, Fig. 17 (b)")
```
